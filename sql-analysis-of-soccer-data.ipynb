```python
import sqlite3 #python
path = "../input/soccer/"  #Insert path here
db = path + 'database.sqlite'
conn = sqlite3.connect(db)
print("Database Connected")
```

```python
#tables from the database
#select * : all the attributes from sqlite_master: database where type = "table"
tables= pd.read_sql("""SELECT name 
                    FROM sqlite_master 
                    where type = 'table';""",conn)
print(tables)
```

```python
q1="""SELECT * FROM Team_Attributes limit 10;"""
r1=pd.read_sql(q1,conn)
print(r1)
```

```python
#where defenceTeamWidth > 50
q2="""SELECT count(team_fifa_api_id) FROM Team_Attributes where (defenceTeamWidth > 50) and (defenceTeamWidthClass = 'Normal') ;"""
r2=pd.read_sql(q2,conn)
print(r2)
```
